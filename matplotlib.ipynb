```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
x = [5, 7, 8, 7, 2, 17]
y = [99, 86, 87, 88, 100, 86]

plt.scatter(x, y, color='green')
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Score")
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [10, 15, 7]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart")
plt.show()
```

```python
ages = [22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79]

plt.hist(ages, bins=7, color='orange', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Programming Language Usage")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
```

```python
plt.plot(x, y, label="Sales")
plt.legend()
```

```python
plt.xlim(0, 10)
plt.ylim(0, 50)
```

```python
plt.subplot(1, 2, 1)  # rows, columns, plot index
plt.plot(x, y)
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.bar(categories, values)
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1, label="y = x^2", linestyle='--', marker='o')
plt.plot(x, y2, label="y = 2^x", linestyle='-', marker='s')

plt.title("Multiple Lines Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```
